```python
folder_list = sorted(glob.glob("/data/kjh/save_log/imagenet_resnet50im/log_imagenet_resnet50im_bs128_ep200_seed_3/*/"))
folder_list= [folder_name for folder_name in folder_list if "layer" in folder_name]

block_n__ = [32, 64, 128, 256, 512]

percent_list = [0.9, 0.8, 0.7] # 남길 값에 대한 비율
for block_n_ in block_n__:
    for layer_path in folder_list:
        for file_name in sorted(os.listdir(layer_path)):
            if ".csv" not in file_name:
                layer_name = layer_path.split("/")[-2]
                csv_file_name = f"{file_name.split('.')[0]}_result_" + f"{block_n_}"+".csv"
                file_path = os.path.join(layer_path, file_name)
                print(layer_name)
                print(file_path)

                with open(file_path, "rb") as f:
                    tensor = pickle.load(f)

                print(tensor.shape)
                oc, ic, h, w = tensor.shape
                if (((oc*h*w)%block_n_)==(block_n_//8)):
                    block_n=block_n_//8
                elif(((oc*h*w)%block_n_)==(block_n_//4)):
                    block_n=block_n_//4
                elif (((oc*h*w)%block_n_)==(block_n_//2)):
                    block_n=block_n_//2
                else:
                    block_n=block_n_

                print(block_n)
                tensor_block_shape = tensor.transpose(1, 0).reshape(ic, (oc*h*w)//block_n, block_n) # oc, ic, h, w -> ic, (oc*h*w) -> ic, ((oc*h*w)/block_n) , block_n

                # ic, (t)/32 , 32
                tensor_block_log2 = torch.floor(torch.log2(torch.abs(tensor_block_shape + 1e-10)))
                tensor_block_log2_max = torch.max(tensor_block_log2, dim=2, keepdims=True)[0]
                tensor_block_log2_min = torch.min(tensor_block_log2, dim=2, keepdims=True)[0]
                tensor_block_log2_mean = torch.mean(tensor_block_log2, dim=2, keepdims=True)[0]
                tensor_block_log2_dist = torch.abs(tensor_block_log2 - tensor_block_log2_mean)

                tensor_block_log2_diff = tensor_block_log2_max - tensor_block_log2_min
                # in_c, t/block_n
                #index, count = tensor_block_log2_diff.flatten().unique(return_counts=True)
                index, count = tensor_block_log2_max.flatten().unique(return_counts=True)
                ori_unique_int = index.numpy().astype(np.int32)
                counts = count.numpy().astype(np.int16)
                diff_arr = np.zeros(60)
                diff_arr[ori_unique_int+50] = count
                result_dict = {"original":diff_arr}

                for percent in percent_list:
                    quantile_index = math.ceil(percent*block_n) # 올림으로 값 retrun
                    print(quantile_index)
                    _, save_indices = torch.topk(tensor_block_log2_dist, quantile_index, dim=2, largest=False)
                    tensor_percent_block_log2 = torch.gather(tensor_block_log2, 2, save_indices)
                    print(tensor_percent_block_log2.shape)
                    tensor_block_log2_max = torch.max(tensor_percent_block_log2, dim=2, keepdims=True)[0]
                    tensor_block_log2_min = torch.min(tensor_percent_block_log2, dim=2, keepdims=True)[0]
                    tensor_block_log2_diff = tensor_block_log2_max - tensor_block_log2_min
                    #index, count = tensor_block_log2_diff.flatten().unique(return_counts=True)
                    index, count = tensor_block_log2_max.flatten().unique(return_counts=True)
                    unique_int = index.numpy().astype(np.int32) 
                    counts = count.numpy().astype(np.int16)
                    #diff_arr = np.zeros(max(ori_unique_int)+1)
                    diff_arr = np.zeros(60)
                    diff_arr[unique_int+50] = count
                    #diff_arr[unique_int] = count
                    result_dict[f'{1-percent:.2f}_prune'] = diff_arr


                # 20% 죽인다. -> 나머지로 우리가 연산을 해야함 / 20% 죽인 값들도 일단 어떻게 디자인 되어있는지 파악은 해보자
                # 20% 를 죽인다 = 32 -> 26 이 되어야 함. 777 66666 55555554444333221
                # blcok 20 % 지움
                # max - min 한 다음 결과 저장        

                result_df = pd.DataFrame(result_dict)
                result_df.to_csv(os.path.join(layer_path, csv_file_name))
```

```python
epoch_num = ["0_","30_","60_","90_","120_","150_","180_"]
file_name_list = ["act_grad","fwd_act","weight_grad","weight"]
file_name_list2 = ["_result","_max","_min"]
group_size = ["","_64","_128","_256","_512"]
final = ".csv"
csv_name_list = []

tt_df_a = pd.DataFrame()
tt_df_b = pd.DataFrame()
tt_df_c = pd.DataFrame()
tt_df_d = pd.DataFrame()
for epoch_num_ in epoch_num:
    for file_name_list_ in file_name_list:
        for file_name_list2_ in file_name_list2:
            for group_size_ in group_size:
                csv_name_list.append(epoch_num_+file_name_list_+file_name_list2_+group_size_+final)
for csv_name in csv_name_list:
    csv_list = sorted(glob.glob("/data/kjh/save_log/imagenet_resnet50im/log_imagenet_resnet50im_bs128_ep200_seed_3/csv/*/"+csv_name))
    
    tt_df_a = pd.DataFrame()
    tt_df_b = pd.DataFrame()
    tt_df_c = pd.DataFrame()
    tt_df_d = pd.DataFrame()
    
    for csv_test in csv_list:
        print(csv_test)
        layer_name = csv_test.split("/")[-2]
        tt_df = pd.read_csv(csv_test, index_col=0)
        concat0 = tt_df["original"]
        concat0.rename(f'original_{layer_name}', inplace=True)
        concat1 = tt_df["0.10_prune"]
        concat1.rename(f'0.10_prune_{layer_name}', inplace=True)
        concat2 = tt_df["0.20_prune"]
        concat2.rename(f'0.20_prune_{layer_name}', inplace=True)   
        concat3 = tt_df["0.30_prune"]
        concat3.rename(f'0.30_prune_{layer_name}', inplace=True)
        
        tt_df_a = pd.concat([tt_df_a,concat0],axis = 1)
        tt_df_b = pd.concat([tt_df_b,concat1],axis = 1)
        tt_df_c = pd.concat([tt_df_c,concat2],axis = 1)
        tt_df_d = pd.concat([tt_df_d,concat3],axis = 1)
    
    if  "min" in csv_test or "max" in csv_test:
        tt_df_a.index -= 50
        tt_df_b.index -= 50
        tt_df_c.index -= 50
        tt_df_d.index -= 50
    tt_df_a.to_csv(os.path.join(csv_path, "0_"+csv_name))
    tt_df_b.to_csv(os.path.join(csv_path, "10_"+csv_name))
    tt_df_c.to_csv(os.path.join(csv_path, "20_"+csv_name))
    tt_df_d.to_csv(os.path.join(csv_path, "30_"+csv_name))
```
